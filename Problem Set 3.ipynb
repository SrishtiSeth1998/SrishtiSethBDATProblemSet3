```python
import pandas as pd

# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 4: Discover the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)

# Step 6: For each occupation, calculate the minimum and maximum ages
min_age_per_occupation = users.groupby('occupation')['age'].min()
max_age_per_occupation = users.groupby('occupation')['age'].max()

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: For each occupation, present the percentage of women and men
gender_percentage_per_occupation = users.groupby(['occupation', 'gender']).size() / users.groupby('occupation').size()

# Display the results
print("Mean Age per Occupation:\n", mean_age_per_occupation)
print("\nMale Ratio per Occupation (Most to Least):\n", male_ratio_per_occupation)
print("\nMinimum Age per Occupation:\n", min_age_per_occupation)
print("\nMaximum Age per Occupation:\n", max_age_per_occupation)
print("\nMean Age per Occupation and Gender:\n", mean_age_per_combination)
print("\nGender Percentage per Occupation:\n", gender_percentage_per_occupation)
```

```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

# Step 4: Select only the Goal column
goals = euro12['Goals']

# Step 5: Number of teams participated
num_teams = len(euro12)

# Step 6: Number of columns
num_columns = euro12.shape[1]

# Step 7: Create the 'discipline' DataFrame
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# Step 8: Sort the teams by Red Cards, then by Yellow Cards
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)

# Step 9: Calculate the mean Yellow Cards per team
mean_yellow_cards = discipline['Yellow Cards'].mean()

# Step 10: Filter teams that scored more than 6 goals
high_scoring_teams = euro12[euro12['Goals'] > 6]

# Step 11: Select teams that start with 'G'
teams_starting_with_G = euro12[euro12['Team'].str.startswith('G')]

# Step 12: Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]

# Step 13: Select all columns except the last 3
all_columns_except_last_3 = euro12.iloc[:, :-3]

# Step 14: Present Shooting Accuracy for specific teams
shooting_accuracy_eng_ita_rus = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]

# Displaying the results
print("Number of teams participated:", num_teams)
print("Number of columns:", num_columns)
print("\nDiscipline DataFrame:\n", discipline_sorted)
print("\nMean Yellow Cards per team:", mean_yellow_cards)
print("\nHigh Scoring Teams:\n", high_scoring_teams)
print("\nTeams starting with 'G':\n", teams_starting_with_G)
print("\nFirst 7 columns:\n", first_7_columns)
print("\nAll columns except last 3:\n", all_columns_except_last_3)
print("\nShooting Accuracy for England, Italy, and Russia:\n", shooting_accuracy_eng_ita_rus)
```

```python
import pandas as pd
import numpy as np

# Step 2: Import the dataset
data = pd.read_csv('wind.txt', delim_whitespace=True, parse_dates=[[0, 1, 2]])

# Step 4: Fix the year 2061
def fix_year(year):
    if year > 2000:
        year -= 100
    return year

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(lambda x: x.replace(year=fix_year(x.year)))

# Step 5: Set datetime index
data.set_index('Yr_Mo_Dy', inplace=True)

# Step 6: Count missing values by location
missing_values = data.isnull().sum()

# Step 7: Count non-missing values
non_missing_count = data.count().sum()

# Step 8: Calculate mean windspeed for the entire dataset
mean_windspeed = data.mean().mean()

# Step 9: Calculate statistics for each location
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

# Step 10: Calculate statistics for each day
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

# Step 11: Calculate average windspeed in January for each location
january_avg = data[data.index.month == 1].mean()

# Step 12: Downsample to yearly frequency
yearly_data = data.resample('Y').mean()

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W').mean()

# Step 15: Calculate statistics for the first 52 weeks
weekly_stats = pd.DataFrame({
    'min': weekly_data.iloc[:52].min(axis=1),
    'max': weekly_data.iloc[:52].max(axis=1),
    'mean': weekly_data.iloc[:52].mean(axis=1),
    'std': weekly_data.iloc[:52].std(axis=1)
})
```

```python
# Print missing values by location
print("Missing Values by Location:")
print(missing_values)

# Print count of non-missing values
print("\nCount of Non-Missing Values:", non_missing_count)

# Print mean windspeed for the entire dataset
print("\nMean Windspeed for the Entire Dataset:", mean_windspeed)

# Print statistics for each location
print("\nLocation Statistics:")
print(loc_stats)

# Print statistics for each day
print("\nDay Statistics:")
print(day_stats)

# Print average windspeed in January for each location
print("\nAverage Windspeed in January for Each Location:")
print(january_avg)

# Print yearly downsampled data
print("\nYearly Downsampled Data:")
print(yearly_data)

# Print monthly downsampled data
print("\nMonthly Downsampled Data:")
print(monthly_data)

# Print weekly downsampled data for the first 52 weeks
print("\nWeekly Downsampled Data (First 52 Weeks):")
print(weekly_data.iloc[:52])

# Print statistics for the first 52 weeks
print("\nWeekly Statistics for First 52 Weeks:")
print(weekly_stats)
```

```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print("First 10 Entries:")
print(chipo.head(10))

# Step 5: Number of observations
num_observations = len(chipo)
print("\nNumber of Observations:", num_observations)

# Step 6: Number of columns
num_columns = chipo.shape[1]
print("\nNumber of Columns:", num_columns)

# Step 7: Print column names
print("\nColumn Names:")
print(chipo.columns)

# Step 8: Indexing information
index_info = chipo.index
print("\nIndexing Information:", index_info)

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("\nMost-Ordered Item:", most_ordered_item)

# Step 10: Quantity of the most-ordered item
most_ordered_quantity = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("\nQuantity of Most-Ordered Item:", most_ordered_quantity)

# Step 11: Most ordered choice_description
most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()
print("\nMost Ordered Choice Description:", most_ordered_choice_description)

# Step 12: Total number of items ordered
total_items_ordered = chipo['quantity'].sum()
print("\nTotal Number of Items Ordered:", total_items_ordered)

# Step 13: Convert item_price to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 14: Total revenue
total_revenue = chipo['item_price'].sum()
print("\nTotal Revenue:", total_revenue)

# Step 15: Number of orders
num_orders = chipo['order_id'].nunique()
print("\nNumber of Orders:", num_orders)

# Step 16: Average revenue per order
avg_revenue_per_order = total_revenue / num_orders
print("\nAverage Revenue Per Order:", avg_revenue_per_order)

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()
print("\nNumber of Different Items Sold:", num_different_items)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Import the dataset
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')


# Step 3: Filter data for the years 1900, 1950, and 2000
years = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years)]

# Step 4: Create the vertical bar chart
plt.figure(figsize=(10, 6))

bar_width = 3.0
opacity = 0.8

plt.bar(filtered_data['Year'] - bar_width/2, filtered_data['Marriages_per_1000'], bar_width, label='Marriages per 1000', alpha=opacity, color='skyblue')
plt.bar(filtered_data['Year'] + bar_width/2, filtered_data['Divorces_per_1000'], bar_width, label='Divorces per 1000', alpha=opacity, color='pink')

# Step 5: Add labels and legend
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks(filtered_data['Year'])
plt.legend()

# Step 6: Display the plot
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Import the data 
data = pd.read_csv('roman-emperor-reigns.csv')
df = pd.DataFrame(data)

# Step 3: Count the number of assassinations and other causes of death
assassination_count = df[df['Cause_of_Death'] == 'Assassinated'].shape[0]
other_count = df.shape[0] - assassination_count

# Step 4: Create the pie chart
labels = ['Assassinated', 'Other']
sizes = [assassination_count, other_count]
colors = ['pink', 'azure']

plt.figure(figsize=(8, 8))

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Step 5: Add title
plt.title('Fraction of Roman Emperors Assassinated')

# Step 6: Display the plot
plt.show()
```
